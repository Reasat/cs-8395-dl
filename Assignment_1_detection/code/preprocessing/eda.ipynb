```python
import os
from glob import glob
from skimage import io
import matplotlib.pyplot as plt
```

```python
dir_data = r'D:\Data\cs-8395-dl\assignment1_data'
```

```python
filepaths_train = glob(os.path.join(dir_data,'train','*.jpg'))
len(filepaths_train)
```

```python
for path in filepaths_train[:3]:
    img = io.imread(path)
    plt.imshow(img)
    plt.show()
    
```

```python
filepaths_train_label = os.path.join(dir_data,'labels','labels.txt')
```

```python
with open(filepaths_train_label,'r') as f:
    label_data = f.readlines()
```

```python
label_dict={}
for data in label_data:
    name, x, y = data.strip().split(' ')
    label_dict[name]=(float(x),float(y))
```

```python
label_dict
```

```python
flnames = label_dict.keys()
```

```python
for path in filepaths_train[:3]:
    img = io.imread(path)
    flname = os.path.basename(path)
    x,y = label_dict[flname]
    plt.imshow(img)
    plt.plot(x*img.shape[1],y*img.shape[0],'r*')
    plt.show()
    
```

```python
col_all=[]
row_all=[]
for path in filepaths_train:
    img = io.imread(path)
    col_all.append(img.shape[1])
    row_all.append(img.shape[0])    
```
