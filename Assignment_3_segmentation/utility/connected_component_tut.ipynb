```python
import cc3d
import numpy as np
# from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
# import matplotlib.pyplot as plt
from plotly import graph_objs as go
```

```python
n=16
labels_in = np.zeros((n, n, n), dtype=np.uint16)

block_1 = np.ones((2,2,2), dtype=np.uint16)
block_2 = np.ones((3,2,2), dtype=np.uint16)

labels_in[4:6,4:6,4:6] = block_1
labels_in[1:4,1:3,1:3] = block_2
```

```python
# np.random.seed(42)
labels_out = cc3d.connected_components(labels_in, 
                                       connectivity=6,
                                       out_dtype=np.uint16)
```

```python
labels_out.shape, np.unique(labels_out.flatten())
```

```python
X, Y, Z = np.mgrid[:n, :n, :n]

fig = go.Figure(data=go.Volume(
    x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
    value=labels_in.flatten(),
    isomin=0.5,
    isomax=0.5,
    opacity=0.2,
    surface_count=21,
    caps= dict(x_show=False, y_show=False, z_show=False), # no caps
    ))
fig.update_layout(scene_camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=0.1, y=2.5, z=0.1)
))

```

```python
fig = go.Figure(data=go.Volume(
    x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
    value=labels_out.flatten(),
    isomin=0.7,
    isomax=0.7,
    opacity=0.2,
    surface_count=21,
    caps= dict(x_show=False, y_show=False, z_show=False), # no caps
    ))
fig.update_layout(scene_camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=0.1, y=2.5, z=0.1)
))

fig.show()
```

```python
# You can extract individual components like so:
N = np.max(labels_out)
for segid in range(1, N+1):
    extracted_image = labels_out * (labels_out == segid)
    print(extracted_image.shape, np.unique(extracted_image))
    fig = go.Figure(data=go.Volume(
    x=X.flatten(), y=Y.flatten(), z=Z.flatten(),
    value=extracted_image.flatten(),
    isomin=0.7,
    isomax=0.7,
    opacity=0.2,
    surface_count=21,
    caps= dict(x_show=False, y_show=False, z_show=False), # no caps
    ))
    fig.update_layout(scene_camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=0.1, y=2.5, z=0.1)
    ))

    fig.show()
```

```python
selem = cube(5)
labels_in_dilated = dilation(labels_in, selem)
```
